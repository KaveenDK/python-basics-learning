```python
class Square:
    def __init__(self, name):
        self.__name = name
    
    def __area(self, side_length):
        return side_length ** 2
    
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name

    def get_area(self, side_length):
        return self.__area(side_length)
    
Square_1 = Square("MySquare")

print(Square_1.get_name())
Square_1.set_name("DK")
print(Square_1.get_name())
print(Square_1.get_area(4))
```
