```python
def calculate_numbers (a,b):
    sum_result = a + b
    diff_result = a - b
    prod_result = a * b
    if b != 0:
        quot_result = a / b
    else:
        quot_result = None
    return sum_result, diff_result, prod_result, quot_result

result = calculate_numbers(10, 2)
print("Sum:", result[0])
print("Difference:", result[1])
print("Product:", result[2])
print("Quotient:", result[3])
```

Python funtion call square_sum to calculate the sum of square numbers from 1 to any given n

```python
def square_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print(square_sum(5))
```

Python Funtion name add_corresponding_numbers that takes two lists of numbers as input and returns a new list containing the sum of correspoding elements from both lists

* if the two list a not equal lenth, you should display and error message

```python
def add_corresponding_numbers(list1, list2):
    if len(list1) != len(list2):
        return "Error: Lists are not of equal length."
    
    new_list = []
    for i in range(len(list1)):
        new_list.append(list1[i] + list2[i])
    return new_list
    
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = add_corresponding_numbers(list1, list2)
print(result)
```

Python Funtion name sum_numbers that accepts any number of numeric arguments and returns there sum

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(1, 2, 3))
print(sum_numbers(10, 20, 30, 40, 50))
```

Python Funtion name average_positive that accept any number of numeric arguments. calculate the average of all positive numbers only.if no positive numbers are provided, the funtion should display no positive numbers provided

```python
def average_positive(*args):
    positive_numbers = [num for num in args if num > 0]
    if not positive_numbers:
        return "No positive numbers provided"
    average = sum(positive_numbers) / len(positive_numbers)
    return average

print(average_positive(1, -2, 3, -4, 5))
print(average_positive(-1, -2, -3))
print(average_positive(-1, 1, 2, 3, 4))
```

```python
def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0:
            total += num
            count += 1

    if count == 0:
        return "No positive numbers provided"
    else:
        return total / count
    
print(average_positive(1, -2, 3, -4, 5))
print(average_positive(-1, -2, -3))
print(average_positive(-1, 1, 2, 3, 4))
```

Python Function name average score that: 

- accept any number of keyword arguments whare 
    * keys are subject name (like math physics)
    * values are scores obtain in each subject 

- calculate and print the average score of all subjects

- if no subjects are provided, then print no subject provided

```python
def average_score(**kwargs):
    if not kwargs:
        print("No subject provided")
        return

    total_score = sum(kwargs.values())
    number_of_subjects = len(kwargs)
    average = total_score / number_of_subjects
    print("Average Score:", average)

average_score(math=75, physics=90, chemistry=85)
average_score(english=80, history=80)
average_score(english=50, history=50, math=50, physics=50, chemistry=0)
average_score()
```

Absolute Value

```python
def absolute_value(num):
    if num < 0:
        return -num
    else:
        return num
    
print(absolute_value(-10))
print(absolute_value(5))
```

Write a Function convert_to_celsius that covers a Fahrenheit temperature to celsius, rounded to two decimal places. Then use the map function to convert the list [32, 68, 77] to celsius and print the result in list 
* celsius = ( fahrenheit - 32 ) * 5 / 9

```python
def convert_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5 / 9
    return celsius

fahrenheit_temps = [32, 68, 77]
celsius_temps = map(convert_to_celsius, fahrenheit_temps)

print(list(celsius_temps))
```

```python
def convert_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5 / 9
    return round(celsius, 2)

fahrenheit_temps = [32, 68, 77]
celsius_temps = map(convert_to_celsius, fahrenheit_temps)

print(list(celsius_temps))
```

```python
# Filter function

list_numbers = [10, 9, 4, 8, 11]

def filter_even_numbers(num):
    return num % 2 == 0

even_numbers = filter(filter_even_numbers, list_numbers)

print(list(even_numbers))
```

Use the Filter Function to extract all numbers From the list [12, 15, 18, 20, 25, 30, 35] that are devisable by 3 but not by 5, and print the result in list

```python
list_numbers = [12, 15, 18, 20, 25, 30, 35]

def find_divisible_by_3_not_5(num):
    return num % 3 == 0 and num % 5 != 0

result = filter(find_divisible_by_3_not_5, list_numbers)
print(list(result))
```
