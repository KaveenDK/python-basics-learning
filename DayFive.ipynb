Write a Python class name Circle with the folowing specification
* and initializer that takes the circle's radiuos as a parameter as stores.
* and instanse method "area" that returns a area of the circle
* and instanse method circumference that returns the circumference of the circle
* and instanse method "display_info" that prints "Circle with radius <radius> has area <area> and circumference <circumference>"

```python
class Circle:
    PI_VALUE = 3.14

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.PI_VALUE * self.radius ** 2

    def circumference(self):
        return 2 * self.PI_VALUE * self.radius

    def display_info(self):
        area = self.area()
        circumference = self.circumference()
        print(f"Circle with radius {self.radius} has area {area} and circumference {circumference}")

circle_1 = Circle(7)
circle_1.display_info()

circle_2 = Circle(14)
circle_2.display_info()
```

```python
from math import pi

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return pi * self.radius ** 2

    def circumference(self):
        return 2 * pi * self.radius

    def display_info(self):
        area = self.area()
        circumference = self.circumference()
        print(f"Circle with radius {self.radius} has area {area} and circumference {circumference}")

circle_1 = Circle(7)
circle_1.display_info()

circle_2 = Circle(14)
circle_2.display_info()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def speak(self):
        print(f"{self.name} says Woof!. It's {self.color} in color")

dog_1 = Dog("Shehan", "Brown")
dog_1.speak()

dog_2 = Dog("Tomi", "Black")
dog_2.speak()
```

create a Python class Shape with a name attribute and a method area that prints "Area calculation not defined for <name>" then create two sub classes Squre and Circle, that inherit from Shape. each subclass should ..
* have its own constructor to initialize appopriate attributes
* overide the area method to calculate and print the area of the shape
* use meaning full instanse attribute for there calculation 

```python
from math import pi

class Shape:
    def __init__(self, name):
        self.name = name

    def area(self):
        print(f"Area calculation not defined for {self.name}")

class Square(Shape):
    def __init__(self, side_length):
        super().__init__("Square")
        self.side_length = side_length

    def area(self):
        area = self.side_length ** 2
        print(f"The area of the {self.name} with side length {self.side_length} is {area}")

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        area = pi * self.radius ** 2
        print(f"The area of the {self.name} with radius {self.radius} is {area}")

square_1 = Square(10)
square_1.area()

circle_1 = Circle(7)
circle_1.area()
```
