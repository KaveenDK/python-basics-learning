# File Handling In Python

```python
file_1 = open("sample_1.txt", "r")
content = file_1.read()
print(content)
file_1.close()
```

```python
with open("sample_1.txt", "r") as file_1:
    content = file_1.read()
    print(content)
```

```python
with open("sample_1.txt", "r") as file_1:
    line_1 = file_1.readline()
    print(line_1)

    line_2 = file_1.readline()
    print(line_2)

    line_3 = file_1.readline()
    print(line_3)

    line_4 = file_1.readline()
    print(line_4)
```

```python
with open("sample_1.txt", "r") as file_1:
    line_1 = file_1.readline()
    line_1 = line_1.strip()
    print(line_1)

    line_2 = file_1.readline()
    line_2 = line_2.strip()
    print(line_2)

    line_3 = file_1.readline()
    line_3 = line_3.strip()
    print(line_3)
```

```python
with open("sample_1.txt", "r") as file_1:
    while True:
        lines = file_1.readline()
        if not lines:
            break
        print(lines.strip())
```

```python
with open("sample_1.txt", "r") as file_1:
    lines = file_1.readlines()
    print(lines)
```

```python
with open("sample_1.txt", "r") as file_1:
    lines = file_1.readlines()

    for line in lines:
        print(line.strip())
```

# "csv" module/library

```python
import csv

with open("person_data.csv", "r", newline='') as csv_file:
    csv_data = csv.reader(csv_file)

    for row in csv_data:
        print(row)
```

```python
import csv

with open("person_data.csv", "r", newline='') as csv_file:
    csv_data = csv.DictReader(csv_file)

    for row in csv_data:
        print(row)
```

```python
import csv

data = [
    {'Name': 'Tharindu', 'Age': '39', 'Country': 'Sri Lanka', 'City': 'Galle'},
    {'Name': 'Alex', 'Age': '16', 'Country': 'Sri Lanka', 'City': 'Colombo'},
    {'Name': 'Mia', 'Age': '17', 'Country': 'Australia', 'City': 'Sydney'},
    {'Name': 'Ravi', 'Age': '15', 'Country': 'USA', 'City': 'New York'},
    {'Name': 'Sara', 'Age': '16', 'Country': 'UK', 'City': 'London'},
    {'Name': 'Liam', 'Age': '17', 'Country': 'Canada', 'City': 'Toronto'}
]

headers = ['Name', 'Age', 'Country', 'City']

with open("sample_csv.csv", "w", newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=headers)

    csv_writer.writeheader()
    csv_writer.writerows(data)

    print("CSV file created successfully.")
```

write a Python code that reads a file call "input.txt" and writes only the even_numbered lines into a new file call "even_lines.txt". (consider the first line as line number one)

```python
with open("input.txt", "r") as input_file:
    lines = input_file.readlines()

even_lines = lines[1::2]

with open("even_lines2.txt", "w") as even_file:
    even_file.writelines(even_lines)
    
```

```python
with open("input.txt", "r") as input_file:
    lines = input_file.readlines()

with open("even_lines.txt", "w") as even_file:
    for i in range(1, len(lines) + 1, 2):
        even_file.write(lines[i])
    
```
