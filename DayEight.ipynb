Q: Fetch a webpage, check if the request was successful, and extract the main heading text from the page's content. You can assume that the base url is "http://example.com"


```python
response = requests.get("https://example.com")

if response.status_code == 200:
    html_content = response.text   
    print(html_content) 

    soup = BeautifulSoup(html_content, 'html.parser')

    main_heading = soup.find('h1')

    if main_heading:
        print(soup.h1.text)
    else:
        print("doesn't have main heading")        
else:
    print("Failed to retrieve HTML content.", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://example.com"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    main_heading = soup.find('h1')
    if main_heading:
        print("Main heading text:", main_heading.text)
    else:
        print("No main heading found.")
```

# API Integration

```python
import requests

url = "https://api.open-meteo.com/v1/forecast"

data = {
    "latitude": 6.0375,
    "longitude": 80.2076,
    "current_weather": True
}

response = requests.get(url, params=data)

if response.status_code == 200:
    weather_data = response.json()
    print("Weather Data:", weather_data)
    print("Current Temperature:", weather_data["current_weather"]["temperature"])
    print("Current Wind Speed:", weather_data["current_weather"]["windspeed"])

else:
    print("Failed to retrieve weather data. Status code:", response.status_code)
```

```python
import requests

url = "https://api.open-meteo.com/v1/forecast"

data = {
    "latitude": 6.0375,
    "longitude": 80.2076,
    "current": "temperature_2m,wind_speed_10m"
}

response = requests.get(url, params=data)

if response.status_code == 200:
    weather_data = response.json()
    print("Weather Data:", weather_data)

    print("Current Temperature:", weather_data["current"]["temperature_2m"])
    print("Current Wind Speed:", weather_data["current"]["wind_speed_10m"])
else:
    print("Failed to retrieve weather data with status code:", response.status_code)
```
