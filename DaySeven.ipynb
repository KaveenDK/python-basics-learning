Boolean Indexing in NumPy

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

condition = arr > 20
print("Condition (arr > 25):", condition)

filtered_arr = arr[condition]
print("Filtered array (elements > 20):", filtered_arr)
```

```python
import numpy as np

arr = np.array([12, 18, 22, 19, 30])

# 15 < arr < 20
condition = (arr > 15) & (arr < 20)
print("Condition (15 < arr < 20):", condition)

filtered_arr = arr[condition]
print("Filtered array (15 < elements < 20):", filtered_arr)

print(arr[(arr > 15) & (arr < 20)])
```

Extract all elements that are 
* Grater than 10 or less than 5
* and not equal to 50

```python
import numpy as np

arr = np.array([
    [12, 5, 7, 20],
    [3, 15, 8, 2],
    [10, 25, 6, 18],
    [4, 11, 30, 9]
])

condition = ((arr > 10) | (arr < 5)) & (arr != 15)
print("Condition ((arr > 10) | (arr < 5)) & (arr != 15):", condition)

filtered_arr = arr[condition]
print("Filtered array (elements > 10 or < 5 and != 15):", filtered_arr)

print("Filtered array using direct indexing:")

arr_filtered = arr [((arr > 10) | (arr < 5)) & (arr != 15)]
print(arr_filtered) 
```

You are given the folowing pandas data frame cpontaning prouduct sales => 

    df = pd.DataFrame({
        "Proudct": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
        "UnitSold": [12, 50, 34, 20, 40],
        "Price": [120000, 1500, 3500, 25000, 6000],
        "Discount": [0.1, 0.05, 0.08, 0.12, 0.7] #discount as a fraction
    })

* calculate the final price after discount
* calculate the total revenue for each product after discount
* for above part 1 and part 2, you should add both result as new coulm to the data frame
* filter and display only the product where revenue is grater than 500000
* calculate the total revenue of all product combine

```python
import pandas as pd

df = pd.DataFrame({
    "Proudct": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold": [12, 50, 34, 20, 40],
    "Price": [120000, 1500, 3500, 25000, 6000],
    "Discount": [0.1, 0.05, 0.08, 0.12, 0.7]
})

# final price after discount
df["FinalPrice"] = df["Price"] * (1 - df["Discount"])
print("Final Price after discount:\n", df)
print(" ")
print(df["FinalPrice"])
print(" ")

# total revenue for each product after discount
df["Revenue"] = df["UnitSold"] * df["FinalPrice"]
print("Total Revenue for each product after discount:\n", df)
print(" ")
print(df["Revenue"])
print(" ")

# filter and display only the product where revenue is greater than 500000
filtered_df = df[df["Revenue"] > 500000]
print("Filtered DataFrame (Revenue > 500000):\n", filtered_df)
print(" ")

# total revenue of all product combined
total_revenue = df["Revenue"].sum()
print("Total Revenue of all products combined:", total_revenue)
```
